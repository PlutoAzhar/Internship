```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
headerTags = [] #empty list to store the headers
for i in soup.find_all(['h1','h2','h3']):
    headerTags.append(i.text)
    
headerTags
```

```python
# Making dataframe

import pandas as pd
z=pd.DataFrame({'HeaderTag':headerTags})
z
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
soup = BeautifulSoup(page.content)
soup
name=[]
for i in soup.find_all('a',class_='restnt-name ellipsis'):
    name.append(i.text)
name
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
cuisine
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
img=[]
for i in soup.find_all('img',class_="no-img"):
    img.append(i.get('data-src'))
img
rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating
print(len(name),len(cuisine),len(location),len(rating),len(img))
h=pd.DataFrame({"RestaurantName":name,'Cuisine':cuisine,'Location':location,'Ratings':rating,'ImageURL':img})
h
```
