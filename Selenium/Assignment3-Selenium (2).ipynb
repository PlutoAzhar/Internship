```python
# Question no. 9:

driver = webdriver.Chrome()
driver.get("https://www.hostelworld.com/")

driver.maximize_window()
time.sleep(5)

search_hotel=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input")
search_hotel.click()

search_hotel.send_keys('London')
time.sleep(5)

search_btn1=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button/div[2]/div[1]")
search_btn1.click()
time.sleep(2)

search_btn2=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button")
search_btn2.click()
time.sleep(5)

#Scraping Hostel Name:

name=[]
a=driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in a:
    b=i.text
    name.append(b)
name=name[2:]


# Scraping Distance from the City:

distance=[]
a=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in a:
    b=i.text
    distance.append(b)    

    
# Scraping Ratings:

rating=[]
a=driver.find_elements(By.XPATH,'//span[@class="number"]')
for i in a:
    b=i.text
    rating.append(b)
rating=rating[2:]

# Scraping Total Review:

comment=[]
a=driver.find_elements(By.XPATH,'//span[@class="keyword"]')
for i in a:
    b=i.text
    comment.append(b)
comment=comment[2:]
comment.insert(12,'None')
# Scraping overall Reiviews:

number=[]
a=driver.find_elements(By.XPATH,'//span[@class="left-margin"]')
for i in a:
    b=i.text
    number.append(b)
    
# Scraping Privates from Price:

alle=[]
a=driver.find_elements(By.XPATH,'//div[@class="property-accommodation-price"]')
for i in a:
    b=i.text
    alle.append(b)
alle=alle[2:]
alle=[w.replace('\n',' to ') for w in alle]
first=alle[0::2]

# Scraping Dorms from Price:

second=alle[1::2]

# Scraping Property Description:

allurl=[]                     # Fetching URL's in allurl
start=0
end=1
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
    for i in url:
        allurl.append(i.get_attribute('href'))
        
descri=[]                     # Fetching Property Description by opening each URL
for i in allurl:
    driver.get(i)
    time.sleep(5)
    try:
        a=driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        descri.append(a.text)
    except NoSuchElementException:
        descri.append('Not Present')        
        
        
# Checking the length of each element:

print(len(name),len(distance),len(rating),len(comment),len(number),len(first),len(second),len(descri))
```
